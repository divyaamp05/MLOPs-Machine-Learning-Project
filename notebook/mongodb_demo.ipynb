```python
import pymongo
from pymongo.errors import OperationFailure
from urllib.parse import quote_plus  # Escape special characters

# Constants
DB_NAME = 'Us_Visa_Project'
COLLECTION_NAME = 'usvisa'
USERNAME = quote_plus('divyaampabattuni')
PASSWORD = quote_plus('Harshu@1003')  # escape @

# Properly escaped connection URL
CONNECTION_URL = f"mongodb+srv://{USERNAME}:{PASSWORD}@cluster0.gnqyggd.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

try:
    # Connect to MongoDB
    client = pymongo.MongoClient(CONNECTION_URL)
    
    print("Connecting to MongoDB...")
    client.server_info()
    print("Connection successful!")

    # Select database and collection
    database = client[DB_NAME]
    collection = database[COLLECTION_NAME]

    # Insert the data
    rec = collection.insert_many(data)
    print("Data inserted successfully. Inserted IDs:", rec.inserted_ids)

except OperationFailure as e:
    print("Authentication failed:", e)
except pymongo.errors.ServerSelectionTimeoutError as err:
    print("Could not connect to MongoDB server:", err)
except Exception as ex:
    print("An error occurred:", ex)
```
